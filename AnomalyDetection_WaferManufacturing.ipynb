```
from pycaret.utils import version
version()
```

```
import pandas as pd
train_data = pd.read_csv('../content/Train.csv')
```

```
train_data
```

```
train_data[train_data['Class'] ==1].shape
```

```
from pycaret.anomaly import *
```

```
#intialize the setup
exp_ano = setup(train_data,pca=True,pca_method='linear',
                log_experiment=True, experiment_name='anomaly_logs')
```

```
models()
```

```
model_iforest = create_model('iforest')
```

```
iforest_results = assign_model(model_iforest)
iforest_results.head()
```

```
# Generating the predictions using Isolation Forest trained model.
iforest_predictions = predict_model(model_iforest, data = train_data)
print(iforest_predictions)
# Checking anomaly rows. Label = 1 is the anomaly data.
iforest_anomaly_rows = iforest_predictions[iforest_predictions['Anomaly'] == 1]
print(iforest_anomaly_rows.head())
# Checking the number of anomaly rows returned by Isolaton Forest.
print(iforest_anomaly_rows.shape) # returned 88 rows
```

```

plot_model(model_iforest, plot='tsne', feature = None, save = False)
```

```
plot_model(model_iforest, plot = 'umap',feature = None, save = False)
```

```
model_tuned = tune_model(model = 'iforest', supervised_target = 'Class',supervised_type ='classification',
              supervised_estimator = 'rf',method='surrogate' ,fold = 10, verbose = True)
```

```
# Generating the predictions using Isolation Forest trained model.
iforest_predictions_tuned = predict_model(model_tuned, data = train_data)
print(iforest_predictions_tuned)
# Checking anomaly rows. Label = 1 is the anomaly data.
iforest_anomaly_rows_tuned = iforest_predictions_tuned[iforest_predictions_tuned['Anomaly'] == 1]
print(iforest_anomaly_rows_tuned.head())
# Checking the number of anomaly rows returned by Isolaton Forest.
print(iforest_anomaly_rows_tuned.shape) # returned 88 rows
```

```
plot_model(model_tuned, plot='tsne', feature = None, save = False)
```
